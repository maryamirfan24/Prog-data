```python
fig = folium.Figure(width=800, height=320) #to set the size of the map on your own
m = folium.Map (
    location=(40.693, -73.9875), zoom_start=16)
folium.Marker (location=(40.693, -73.9875),
               tooltip = "hello!",
               popup="where the magic happens!",
               icon=folium.Icon(icon="star", color="blue")
               ).add_to(m) #add a marker for each point that you want to mark
fig.add_child(m)
```

```python
fig = folium.Figure(width=800, height=320) #to set the size of the map on your own
m = folium.Map (location=(40.693, -73.9875), zoom_start=16)

#iterate over our dataset
#for each row, add a market at the location of that tree

for idx, item in df.sample(1000).iterrows(): 
    print(item)

    folium.Marker (location=(item['latitude'], item['longitude']),
              tooltip = str(item['tree_id']),
               popup=item['health'],
               icon=folium.Icon(icon="star", color="green")
               ).add_to(m) #add a marker for each point that you want to mark
fig.add_child(m)
```

```python
#code to limit analysis to downtown brooklyn
fig = folium.Figure(width=800, height=320) #to set the size of the map on your own
m = folium.Map (location=(40.693, -73.9875), zoom_start=16)

#iterate over our dataset
#for each row, add a market at the location of that tree

for idx, item in downtown_df.iterrows(): 
    print(item)

    folium.Marker (location=(item['latitude'], item['longitude']),
              tooltip = str(item['tree_id']),
               popup=item['health'],
               icon=folium.Icon(icon="star", color="green")
               ).add_to(m) #add a marker for each point that you want to mark
    
fig.add_child(m)
```

```python
fig = folium.Figure(width=800, height=320) #to set the size of the map on your own
m = folium.Map (location=(40.7128, -74.0061), zoom_start=14, prefer_canvas=True)

#iterate over our dataset
#for each row, add a market at the location of that tree

for idx, item in df.sample(1000).iterrows(): 
    print(item)

    folium.Circle (location=(item['latitude'], item['longitude']),
              radius = 6,
               fill = True,
               opacity =0.5,
               color = 'blue'
               ).add_to(m) 
    
fig.add_child(m)
```
